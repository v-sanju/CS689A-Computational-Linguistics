### Pre Processing

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
# for reading a file
def read_file(path):
    data_set=[]
    with open(path, 'r', encoding='utf-8') as file:
        file_contents = file.read()
    file_contents = file_contents.split('\n')
    return file_contents

# for writing a data to a file
def write_to_file(file_path, string_list):
    with open(file_path, 'w', encoding='utf-8') as file:
        for string in string_list:
            file.write(string + '\n')  # Write the string followed by a newline character
```

```python
# read files of 3 different languages (1 english, 1 mother tongue, 1 any other indic lang)
hindi = read_file('/kaggle/input/test-dataset-3/test.hi')
english = read_file('/kaggle/input/test-dataset-3/test.en')
gujrati = read_file('/kaggle/input/test-dataset-3/test.gu')
```

```python
# generate random indices for 1000 random lines from each of the dataset
import random
random.seed(100)
random_indices = random.sample(range(len(hindi)), 1000) 
```

```python
# generated data of corresponding 1000 random lines from all 3 languages
rand_hin = [hindi[i] for i in random_indices]
rand_eng = [english[i] for i in random_indices]
rand_guj = [gujrati[i] for i in random_indices]
```

```python
# generated txt files for above data
write_to_file('/kaggle/working/random_hin_1000.txt', rand_hin)
write_to_file('/kaggle/working/random_eng_1000.txt', rand_eng)
write_to_file('/kaggle/working/random_guj_1000.txt', rand_guj)
```

## NLLB-200 Model

```python
# Function for translation
def nllb_translate_to_lang(lang,src_lang, targ_lang):
    tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M", src_lang=src_lang
    )
    model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
    translated=[]
    for i in range(1000):
        inputs = tokenizer(lang[i] , return_tensors="pt")

        translated_tokens = model.generate(
            **inputs, forced_bos_token_id=tokenizer.lang_code_to_id[targ_lang], max_length=30
        )
        translated.append(tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0])
    return translated
```

### English to Hindi and Hindi to English translation

```python
eng_to_hin = nllb_translate_to_lang(rand_eng,'eng_Latn', 'hin_Deva')
write_to_file('/kaggle/working/nllb_translated_eng_to_hin.txt', eng_to_hin)

hin_to_eng = nllb_translate_to_lang(rand_eng, 'hin_Deva','eng_Latn')
write_to_file('/kaggle/working/nllb_translated_hin_to_eng.txt', hin_to_eng)
```

### Hindi to Gujrati and Gujrati to Hindi Translation

```python
guj_to_hin = nllb_translate_to_lang(rand_eng,'guj_Gujr', 'hin_Deva')
write_to_file('/kaggle/working/nllb_translated_guj_to_hin.txt', guj_to_hin)

hin_to_guj = nllb_translate_to_lang(rand_eng, 'hin_Deva', 'guj_Gujr')
write_to_file('/kaggle/working/nllb_translated_hin_to_guj.txt', hin_to_guj)
```
